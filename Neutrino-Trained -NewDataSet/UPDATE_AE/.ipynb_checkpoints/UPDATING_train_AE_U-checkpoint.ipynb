```python
wireplane = "U"
np.random.seed(42)
path = '../processed_data/current/'
filenames = [path+f for f in listdir(path) if (isfile(join(path, f)) and wireplane in f)]
filenames
```

```python
x_test = np.load('../processed_data/current/x_test_U.npy')
x_train = np.load('../processed_data/current/x_train_U.npy')
y_test = np.load('../processed_data/current/y_test_AE_U.npy')
y_train = np.load('../processed_data/current/y_train_AE_U.npy')
```

```python
for idx in range(4,6):
    fig, (ax1) = plt.subplots(figsize=(12, 2), facecolor='white')
    ax1.plot(x_train[idx],color='blue', label='signal+noise')
    ax1.plot(y_train[idx],color='m',alpha=0.7, label='target')
    ax1.set_title("waveform", fontsize=15)
    ax1.legend()

    plt.show()
```

```python
mean = x_train.mean()
std = x_train.std()
#np.save("../latest_models/mean_AE_"+wireplane+"_nu", mean)
#np.save("../latest_models/scale_AE_"+wireplane+"_nu", std)
```

```python
x_train_scaled = (x_train-mean)/std
x_test_scaled = (x_test-mean)/std
```

```python
y_train_scaled = []
for i in range(len(y_train)):
    if sum(y_train[i]) == 0:
        y_train_scaled.append(y_train[i])
    else:
        scaled = (y_train[i]-mean)/std
        y_train_scaled.append(scaled)
y_train_scaled = np.array(y_train_scaled)
```

```python
y_test_scaled = []
for i in range(len(y_test)):
    if sum(y_test[i]) == 0:
        y_test_scaled.append(y_test[i])
    else:
        scaled = (y_test[i]-mean)/std
        y_test_scaled.append(scaled)
y_test_scaled = np.array(y_test_scaled)
```

```python
for idx in range(4,6):
    fig, (ax1) = plt.subplots(figsize=(12, 2), facecolor='white')
    ax1.plot(x_train_scaled[idx],color='blue', label='signal+noise')
    ax1.plot(y_train_scaled[idx],color='m',alpha=0.7, label='target')
    ax1.set_title("waveform", fontsize=15)
    ax1.legend()
    print(sum(y_train_scaled[idx]))
    plt.show()
```

```python
y_true_rescaled = []

#print('rescalling y_true...')
for i in tqdm.trange(len(y_train_scaled)):
    if sum(y_train_scaled[i]) == 0:
        y_true_rescaled.append(y_train_scaled[i])
    else:
        y_true_rescaled.append((std*y_train_scaled[i]+mean))
y_true_rescaled = np.array(y_true_rescaled)

sig_ranges = []
#print('finding ranges where there are signals: ') 
for i in tqdm.trange(len(y_true_rescaled)):
    wave =  y_true_rescaled[i]
    sig_ranges.append(funcs.merge_ranges(wave, 5))
sig_ranges = np.array(sig_ranges, dtype=object)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalAveragePooling1D, Dropout, Dense,Flatten, AveragePooling1D
from tensorflow.keras.layers import Input,  UpSampling1D
from tensorflow.keras.models import Model
#from keras import backend as K
```

```python
from tensorflow.keras.models import load_model
model = load_model('../latest_models/model_' + wireplane + 'plane_nu.h5')
```

```python
for i,layer in enumerate(model.layers):
    print(i,layer.name)
```

```python
#del x
time_periods = 200
input_wave = Input(shape=(x_train_scaled.shape[1], 1)) 
x = Conv1D(filters=16, kernel_size=3, strides=2, activation = "relu", weights=model.layers[0].get_weights() , input_shape=(time_periods,1))(input_wave)
x = MaxPooling1D(pool_size=2)(x)
 
#second convolutional block
x = Conv1D(filters=32, kernel_size=5, strides=2, weights=model.layers[2].get_weights(), activation = "relu", )(x)
x = MaxPooling1D(pool_size=2)(x)

encoded = Conv1D(filters=64, kernel_size=9, weights=model.layers[5].get_weights(), activation = "relu")(x)


x = Conv1D(filters=64, kernel_size=9, padding = "same", activation = "relu")(encoded)
x = UpSampling1D(4)(x)

x = Conv1D(filters=32, kernel_size=5, padding = "same", activation = "relu")(x)
x = UpSampling1D(3)(x)

x = Conv1D(filters=16, kernel_size=3,  activation = "relu")(x)
x = UpSampling1D(6)(x)

decoded = Conv1D(filters=1, kernel_size=5,  activation = "linear")(x)



autoencoder = Model(input_wave, decoded)
```

```python
from tensorflow.keras.callbacks import Callback
class CustomLossCallback(Callback):
    def __init__(self, additional_array):
        super(CustomLossCallback, self).__init__()
        self.additional_array = additional_array

    def on_train_batch_begin(self, batch, logs=None):
        # Modify the model's loss using the additional array
        logs = logs or {}
        logs['loss'] *= tf.reduce_mean(self.additional_array)

# Create the custom callback
custom_callback = CustomLossCallback(sig_ranges)
```

```python
# Define the custom loss function
import tensorflow.keras.backend as K


def custom_loss(mask):
    def loss(y_true, y_pred):
        # Apply mask to y_true and y_pred
        y_true_masked = y_true * mask
        y_pred_masked = y_pred * mask
        
        # Compute the mean squared error
        mse = K.mean(K.square(y_true_masked - y_pred_masked))
        
        return mse
    return loss

mask = np.random.rand(*y_train_scaled.shape)


# train/val split
x_train_scaled, x_val_scaled,y_train_scaled, y_val_scaled, mask, mask_val  = train_test_split(
    x_train_scaled, y_train_scaled, mask, test_size=0.2, shuffle=False
)

```

```python
print(len(x_train_scaled))
```

```python
# Compile the model with the custom loss function and additional array
autoencoder.compile(optimizer='adam', loss=custom_loss(mask))

# Define the number of epochs and batch size
epochs = 100
batch_size = 2048

# Define lists to track the loss and validation loss
train_loss = []
val_loss = []

```

```python
# Define the training loop
for epoch in range(epochs):
    # Shuffle the training data
    indices = np.arange(x_train_scaled.shape[0])
    np.random.shuffle(indices)
    
    # Iterate over batches of training data
    for i in range(0, x_train_scaled.shape[0], batch_size):
        # Select a batch of training data
        x_batch = x_train_scaled[indices[i:i+batch_size]]
        y_batch = y_train_scaled[indices[i:i+batch_size]]
        
        # Train the model on the batch
        loss = autoencoder.train_on_batch(x_batch, y_batch)
    
    # Compute the training loss for the epoch
    train_loss_epoch = autoencoder.evaluate(x_train_scaled, y_train_scaled, verbose=0)
    train_loss.append(train_loss_epoch)
    
    # Compute the validation loss for the epoch
    val_loss_epoch = autoencoder.evaluate(x_val_scaled, y_val_scaled, verbose=0)
    val_loss.append(val_loss_epoch)
    
    # Print the loss and validation loss for the epoch
    print("Epoch: {}, Loss: {}, Val Loss: {}".format(epoch, train_loss_epoch, val_loss_epoch))
```
