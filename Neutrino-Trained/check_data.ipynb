```python
import sys
import numpy as np
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
```

```python
wireplane = 'Z'
path = './processed_data/'
```

```python
x_train = np.load(path+'x_train_' + wireplane + '.npy')
x_test = np.load(path+'x_test_' + wireplane + '.npy')
y_train_AE = np.load(path+'y_train_AE_' + wireplane + '.npy')
y_test_AE = np.load(path+'y_test_AE_' + wireplane + '.npy')

y_train_ROI = np.load(path+'y_train_ROI_' + wireplane + '.npy')
y_test_ROI = np.load(path+'y_test_ROI_' + wireplane + '.npy')
```

```python
for i in range(10):
    fig = plt.figure(figsize=(10,4))
    plt.plot(x_train[i], color='black', alpha=0.3, label='raw_train_data')
    plt.plot(y_train_AE[i], color='blue', alpha=1, label='clean_train_data')
    plt.title('ROI truth: '  + str(y_train_ROI[i]))
    plt.legend()
    plt.show()
```
