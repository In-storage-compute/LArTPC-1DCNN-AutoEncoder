```python
signal_dataset_split = []
for signal in signal_data:
    roi_truth_signals = np.ones(signal.shape[0])
    _x_train, _x_test, _y_train, _y_test = train_test_split(signal, roi_truth_signals, test_size=0.2, shuffle=True)
    _x_test, _x_valid, _y_test, _y_valid = train_test_split(_x_test, _y_test, test_size = 0.5, shuffle=True)
    print(_x_train.shape, _x_test.shape, _x_valid.shape)
    signal_dataset_split.append([[_x_train, _y_train], [_x_test, _y_test], [_x_valid, _y_valid]])
    
```

```python
noise_dataset_split = []
for noise in noise_data:
    roi_truth_noise = np.zeros(noise.shape[0])
    _x_train, _x_test, _y_train, _y_test = train_test_split(signal, roi_truth_noise, test_size=0.2, shuffle=True)
    _x_test, _x_valid, _y_test, _y_valid = train_test_split(_x_test, _y_test, test_size = 0.5, shuffle=True)
    print(_x_train.shape, _x_test.shape, _x_valid.shape)
    noise_dataset_split.append([[_x_train, _y_train], [_x_test, _y_test], [_x_valid, _y_valid]])
    
```

```python
train_dataset = []
valid_dataset = []
test_dataset = []
for i, signal_set in enumerate(signal_dataset_split):
    train_sig = signal_set[0]
    test_sig = signal_set[1]
    val_sig = signal_set[2]
    
    train_noise = noise_dataset_split[i][0]
    test_noise = noise_dataset_split[i][1]
    val_noise = noise_dataset_split[i][2]
    
    X_train = np.concatenate((train_sig[0], train_noise[0]))
    Y_train = np.concatenate((train_sig[1], train_noise[1]))
    
    for i in range(10):
        X_train, Y_train = shuffle(X_train, Y_train)
    
    train_dataset.append([X_train, Y_train])
        
    X_valid = np.concatenate((val_sig[0], val_noise[0]))
    Y_valid = np.concatenate((val_sig[1], val_noise[1]))
    
    for i in range(10):
        X_valid, Y_valid = shuffle(X_valid, Y_valid)
    
    valid_dataset.append([X_valid, Y_valid])
        
    X_test = np.concatenate((test_sig[0], test_noise[0]))
    Y_test = np.concatenate((test_sig[1], test_noise[1]))
    
    for i in range(10):
        X_test, Y_test = shuffle(X_test, Y_test)
    
    test_dataset.append([X_test, Y_test])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalMaxPooling1D, \
    Dropout, Dense,Flatten, AveragePooling1D, BatchNormalization
```

```python
time_periods = 200
def create_model():
    model = Sequential()  
    #first convolutional block
    model.add(Conv1D(filters=16, kernel_size=3, strides=2, activation = "relu", input_shape=(time_periods,1)))
    model.add(MaxPooling1D(pool_size=2))

    #second convolutional block
    model.add(Conv1D(filters=32, kernel_size=5, strides=2, activation = "relu"))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.1))

    #forth convolutional block                                      
    model.add(Conv1D(filters=64, kernel_size=9, activation = "relu"))
    model.add(GlobalMaxPooling1D())
    model.add(Dropout(0.2)) 

    model.add(Flatten())



    model.add(Dense(1, activation="sigmoid", name="wavrec_out"))
    return model
```

```python
model = create_model()

#lr=0.001
adam = tf.keras.optimizers.legacy.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

model.compile(optimizer = adam, loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
earlystop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=10,
    verbose=0,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
)
```

```python
full_history = []
for i in range(len(train_dataset)):
    print('Loop:', i, 'of', 18)
    
    x_train_scaled = (train_dataset[i][0]-mean)/std
    y_train = train_dataset[i][1]
    x_valid_scaled = (valid_dataset[i][0]-mean)/std
    y_valid = valid_dataset[i][1]
    
    if i == 3:
        append_train_X = (train_dataset[17][0]-mean)/std
        append_train_X = append_train_X[0:24000]
        append_train_Y = train_dataset[17][1]
        append_train_Y = append_train_Y[0:24000]
        
        append_valid_X = (valid_dataset[17][0]-mean)/std
        append_valid_X = append_valid_X[0:3000]
        append_valid_Y = valid_dataset[17][1]
        append_valid_Y = append_valid_Y[0:3000]
        
        x_train_scaled = np.concatenate((x_train_scaled, append_train_X))
        y_train = np.concatenate((y_train, append_train_Y))
        
        x_valid_scaled = np.concatenate((x_valid_scaled, append_valid_X))
        y_valid = np.concatenate((y_valid, append_valid_Y))
        
        for i in range(5):
            x_train_scaled, y_train = shuffle(x_train_scaled, y_train)
            x_valid_scaled, y_valid = shuffle(x_valid_scaled, y_valid)
            
        if i == 7:
            append_train_X = (train_dataset[16][0]-mean)/std
            append_train_X = append_train_X[0:24000]
            append_train_Y = train_dataset[16][1]
            append_train_Y = append_train_Y[0:24000]

            append_valid_X = (valid_dataset[16][0]-mean)/std
            append_valid_X = append_valid_X[0:3000]
            append_valid_Y = valid_dataset[16][1]
            append_valid_Y = append_valid_Y[0:3000]

            x_train_scaled = np.concatenate((x_train_scaled, append_train_X))
            y_train = np.concatenate((y_train, append_train_Y))

            x_valid_scaled = np.concatenate((x_valid_scaled, append_valid_X))
            y_valid = np.concatenate((y_valid, append_valid_Y))

            for i in range(5):
                x_train_scaled, y_train = shuffle(x_train_scaled, y_train)
                x_valid_scaled, y_valid = shuffle(x_valid_scaled, y_valid)

        if i == 11:
            append_train_X = (train_dataset[17][0]-mean)/std
            append_train_X = append_train_X[-24000:]
            append_train_Y = train_dataset[17][1]
            append_train_Y = append_train_Y[-24000]

            append_valid_X = (valid_dataset[17][0]-mean)/std
            append_valid_X = append_valid_X[-3000:]
            append_valid_Y = valid_dataset[17][1]
            append_valid_Y = append_valid_Y[-3000:]

            x_train_scaled = np.concatenate((x_train_scaled, append_train_X))
            y_train = np.concatenate((y_train, append_train_Y))

            x_valid_scaled = np.concatenate((x_valid_scaled, append_valid_X))
            y_valid = np.concatenate((y_valid, append_valid_Y))

            for i in range(5):
                x_train_scaled, y_train = shuffle(x_train_scaled, y_train)
                x_valid_scaled, y_valid = shuffle(x_valid_scaled, y_valid)

        if i == 17:
            append_train_X = (train_dataset[0][0]-mean)/std
            #append_train_X = append_train_X[-48000:]
            append_train_Y = train_dataset[0][1]
            #append_train_Y = append_train_Y[-48000]

            append_valid_X = (valid_dataset[0][0]-mean)/std
            #append_valid_X = append_valid_X[-6000:]
            append_valid_Y = valid_dataset[0][1]
            #append_valid_Y = append_valid_Y[-6000:]

            x_train_scaled = np.concatenate((x_train_scaled, append_train_X))
            y_train = np.concatenate((y_train, append_train_Y))

            x_valid_scaled = np.concatenate((x_valid_scaled, append_valid_X))
            y_valid = np.concatenate((y_valid, append_valid_Y))

            for i in range(5):
                x_train_scaled, y_train = shuffle(x_train_scaled, y_train)
                x_valid_scaled, y_valid = shuffle(x_valid_scaled, y_valid)
    
        
        
    
    x_train_scaled = x_train_scaled.reshape(x_train_scaled.shape[0], x_train_scaled.shape[1], 1)
    x_valid_scaled = x_valid_scaled.reshape(x_valid_scaled.shape[0], x_valid_scaled.shape[1], 1)
    
    history = model.fit(x_train_scaled,                                                              
                    y_train,
                    epochs=100,      
                   # epochs=30, 
                    #epochs = 6,
                    callbacks=[earlystop],
                     # callbacks=callbacks_list,                         
                    validation_data=(x_valid_scaled, y_valid),                                               
                    verbose=1)
    
    
    full_history.append(history)
    
```
